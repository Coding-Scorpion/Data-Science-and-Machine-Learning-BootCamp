```python
# Create sample data with column names and additional data at the beginning
data = """BEGINNING OF THE FILE
% Day, Month, Year, Skip, Avg Power
01, 01, 2000, x876, 13 % crazy day!
% we dont have Jan 03rd
04, 01, 2000, xfed, 55
"""

# Write the data to a text file
with open('Data.txt', 'w') as file:
    file.write(data)
```

```python
# loadtxt() automatically generates an array from the txt file
arr = np.loadtxt('Data.txt',
                 skiprows=1,         # Skip the first row (header) of the text file
                 dtype=int,          # Data type of the loaded array elements (integer)
                 delimiter=",",      # Delimiter used in the text file (comma-separated)
                 usecols=(0, 1, 2, 4), # Load only columns 0, 1, 2, and 4 from the text file
                 comments="%")       # Ignore lines in the text file that start with '%'

# Save an array into a txt file
np.savetxt('array_data.txt', arr)  # Save the array 'arr' into a new text file named 'filename.txt'
```

### Broadcasting

The term broadcasting describes how NumPy treats arrays with different shapes during arithmetic operations.

The smaller array is broadcast across the larger array so that they have compatible shapes.

#### Broadcasting Rules

**1. Make the two arrays have the same number of dimensions.**<br>
- If the numbers of dimensions of the two arrays are different, add new dimensions with size 1 to the head of the array with the smaller dimension.<br>

Means agar 1 array ki shape (3,4) ha oor second array ki shape (4,) ha to, jo smaller array ha uss k head pr new dimension add krni ha jiss ka size 1 hona chaye .... so after broadcasting (4,) ---- (1,4) ma change ho jay ga <br>

**2. Make each dimension of the two arrays the same size.**<br>
- If the sizes of each dimension of the two arrays do not match, dimensions with size 1 are stretched to the size of the other array.<br>
Means jis array ki dimension 1 ho ge uss ko stretched kr k dosry array ki dimension k equal krna ha...<br>

first array ki shape (3,4)  <br>
second array ki shape (1,4) ha to 1 ko 3 tak increase kr dyn gy ...<br>

- If there is a dimension whose size is not 1 in either of the two arrays, it cannot be broadcasted, and an error is raised.<br>

means agar dono array ma sa kisi b array ki kisi b dimension ka size 1 nehe ho ga to broadcasting nehe ho ge<br>

2 arrays of shape (3,4) and (4,6) broadcast nehe ho skty q k dono ma sa kisi array ki b dimension ka size 1 nehe ha<br>

```python

a = np.arange(12).reshape(4,3)
b = np.arange(3)

print(a+b)
```

```python
a = np.array([1])
# shape -> (1,1)
b = np.arange(4).reshape(2,2)
# shape -> (2,2)

print(a)
print(b)

print(a+b)
```

```python
import matplotlib.pyplot as plt
a = np.linspace(0, 1, 15) - 0.5
b = a[:, np.newaxis] # b.shape == (15, 1)
dist2 = a**2 + b**2 # broadcasting sum.
dist = np.sqrt(dist2)
plt.imshow(dist)
plt.colorbar()
```
